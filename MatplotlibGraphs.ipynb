Certainly, here are some common statistical graph types using Python's Matplotlib library, along with example code for each:

1. Line Plot

Purpose: To visualize trends and relationships between two variables.
Example:
Python

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 7]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
2. Scatter Plot

Purpose: To visualize the relationship between two variables, showing individual data points.
Example:
Python

import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = 2 * x + 1 + np.random.randn(50) 

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
3. Bar Chart

Purpose: To compare categorical data, showing the frequency or magnitude of different categories.
Example:
Python

import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 15, 8, 12]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
4. Histogram

Purpose: To visualize the distribution of a single numerical variable.
Example:
Python

import matplotlib.pyplot as plt
import numpy as np

data = np.random.normal(loc=50, scale=10, size=1000) 

plt.hist(data, bins=20)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
5. Box Plot

Purpose: To visualize the distribution of a dataset, including quartiles, median, and outliers.
Example:
Python

import matplotlib.pyplot as plt
import numpy as np

data = [np.random.normal(loc=i, scale=1, size=100) for i in range(1, 4)]

plt.boxplot(data)
plt.xticks([1, 2, 3], ['Group 1', 'Group 2', 'Group 3'])
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
6. Pie Chart

Purpose: To visualize the proportion of different categories within a whole.
Example:
Python

import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.show()
7. Heatmap

Purpose: To visualize the relationship between two variables in a matrix format, often used to represent correlation matrices.
Example:
Python

import matplotlib.pyplot as plt
import numpy as np

data = np.random.rand(5, 5)

plt.imshow(data, cmap='viridis')
plt.colorbar()
plt.title('Heatmap')
plt.show()
8. Scatter Matrix (Pairplot)

Purpose: To visualize the relationships between multiple variables in a dataset.
Example:
Python

import seaborn as sns
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(100, 4), columns=['A', 'B', 'C', 'D'])
sns.pairplot(df)
plt.show()

# Line Plot
# Purpose: To visualize trends and relationships between two variables.

```python
# Line Plot
# Purpose: To visualize trends and relationships between two variables. Example: Python

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5] 
y = [2, 4, 5, 4, 7]

plt.plot(x, y) 
plt.xlabel('X-axis') 
plt.ylabel('Y-axis') 
plt.title('Line Plot') 
plt.show() 
```

# Scatter Plot

# Purpose: To visualize the relationship between two variables, showing individual data points. Example: Python

```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.random.rand(50) 
y = 2 * x + 1 + np.random.randn(50)

plt.scatter(x, y) 
plt.xlabel('X-axis') 
plt.ylabel('Y-axis') 
plt.title('Scatter Plot') 
plt.show() 
```

# Bar Chart

# Purpose: To compare categorical data, showing the frequency or magnitude of different categories. Example: Python



```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D'] 
values = [10, 15, 8, 12]

plt.bar(categories, values) 
plt.xlabel('Categories') 
plt.ylabel('Values') 
plt.title('Bar Chart') 
plt.show() 
```

# Histogram

# Purpose: To visualize the distribution of a single numerical variable. 



```python
import matplotlib.pyplot as plt 
import numpy as np

data = np.random.normal(loc=50, scale=10, size=1000)

plt.hist(data, bins=20) 
plt.xlabel('Value') 
plt.ylabel('Frequency') 
plt.title('Histogram') 
plt.show() 
```

# Box Plot

# Purpose: To visualize the distribution of a dataset, including quartiles, median, and outliers.

```python
import matplotlib.pyplot as plt 
import numpy as np

data = [np.random.normal(loc=i, scale=1, size=100) 
        for i in range(1, 4)]

plt.boxplot(data) 
plt.xticks([1, 2, 3], ['Group 1', 'Group 2', 'Group 3']) 
plt.ylabel('Values') 
plt.title('Box Plot') 
plt.show()
```

# Pie Chart

# Purpose: To visualize the proportion of different categories within a whole. Example: Python



```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D'] 
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90) 
plt.title('Pie Chart') 
plt.show() 
```


#Heatmap

# Purpose: To visualize the relationship between two variables in a matrix format, often used to represent correlation matrices. Example: Python



```python
import matplotlib.pyplot as plt 
import numpy as np

data = np.random.rand(5, 5)

plt.imshow(data, cmap='viridis') 
plt.colorbar() 
plt.title('Heatmap') 
plt.show() 

```

# Scatter Matrix (Pairplot)

# Purpose: To visualize the relationships between multiple variables in a dataset.



```python
import seaborn as sns 
import pandas as pd 
import numpy as np

df = pd.DataFrame(np.random.randn(100, 4), columns=['A', 'B', 'C', 'D']) 
sns.pairplot(df) 
plt.show() 
```
