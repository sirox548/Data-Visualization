```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Inline flag uses approriate backend to make figures appear inline in tebook 
%matplotlib inline
```

# Loading CSV or XLSX File

```python
# Loading car loan data from csv file

filename = 'Files/table_sienna.csv'
df = pd.read_csv(filename)

#df = pd.read_excel(filename)
```

```python
df.head()
```

```python
# Labels for figure
month_number = df.loc[:, 'Month'].values
interest_paid = df.loc[:, 'interest_paid'].values
principal_paid = df.loc[:, 'principal_paid'].values
```

```python
month_number
```

## Editing Graph size, colors and marker

```python
plt.plot(month_number, interest_paid,  c='k', marker = '.', markersize = 10)
plt.plot(month_number, principal_paid, c='b', marker = ',', markersize = 10 )
```

```python
plt.plot(month_number, interest_paid,  c='b', marker = '.', markersize = 10)
plt.plot(month_number, principal_paid, c='k', marker = ',', markersize = 10 )
```

## Matlab-Style vs Object Oriented Syntax

```python
# MATLAB STYLE
plt.plot(month_number, interest_paid,  c='k')
plt.plot(month_number, principal_paid, c='b')
```

```python
# OBJECT-ORIENTED
# tuple unpacking
x, y = (3, 9)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(month_number, interest_paid,  c='k')
axes.plot(month_number, principal_paid,  c='r')
```

## Adding Lables,Titles and Limits

```python
# SET xLim and yLim
plt.plot(month_number, interest_paid,  c='r')
plt.plot(month_number, principal_paid, c='b')

# LIMITS
plt.xlim(left = 1, right=70)
plt.ylim(bottom =0, top=800)

#LABELS
plt.xlabel('Months')
plt.ylabel('Dollars')
plt.title('Interest and Principal Paid per Month')
```

```python
# Font Adjustment
plt.plot(month_number, interest_paid,  c='r')
plt.plot(month_number, principal_paid, c='b')

# LIMITS
plt.xlim(left = 1, right=70)
plt.ylim(bottom =0, top=800)

#LABELS
plt.xlabel('Months', fontsize=14)
plt.ylabel('Dollars', fontsize=14)
plt.title('Interest and Principal Paid per Month', fontsize=14)

# Number size Adjustment
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
```

## Object Oriented style editing

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(month_number, interest_paid,  c='k')
axes.plot(month_number, principal_paid,  c='r')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(month_number, interest_paid,  c='k')
axes.plot(month_number, principal_paid,  c='r')

# Labels
axes.set_xlabel('Month', fontsize = 14)
axes.set_ylabel('Dollars',  fontsize = 14)
axes.set_title('Interest and Principal paid per month.',  fontsize = 15)

# Ticks
axes.tick_params(axis = 'x', labelsize = 13)
axes.tick_params(axis = 'y', labelsize = 13)
```

## Adding Legends

```python
# Font Adjustment
plt.plot(month_number, interest_paid,  c='r', label= 'Interest')
plt.plot(month_number, principal_paid, c='b', label= 'Principal')


# LIMITS
plt.xlim(left = 1, right=70)
plt.ylim(bottom =0, top=800)

#LABELS
plt.xlabel('Months', fontsize=14)
plt.ylabel('Dollars', fontsize=14)
plt.title('Interest and Principal Paid per Month', fontsize=14)

# Number size Adjustment
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)


# Legends
plt.legend(loc=(1.02,0))
```

```python
# Font Adjustment
plt.plot(month_number, interest_paid,  c='r', label= 'Interest')
plt.plot(month_number, principal_paid, c='b', label= 'Principal')


# LIMITS
plt.xlim(left = 1, right=70)
plt.ylim(bottom =0, top=800)

#LABELS
plt.xlabel('Months', fontsize=14)
plt.ylabel('Dollars', fontsize=14)
plt.title('Interest and Principal Paid per Month', fontsize=14)

# Number size Adjustment
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)


# Legends
plt.legend(loc='center right')
```

## Saving images or graphs

```python
# Style
plt.style.use('seaborn')
# Font Adjustment
plt.plot(month_number, interest_paid,  c='r', label= 'Interest')
plt.plot(month_number, principal_paid, c='b', label= 'Principal')


# LIMITS
plt.xlim(left = 1, right=70)
plt.ylim(bottom =0, top=800)

#LABELS
plt.xlabel('Months', fontsize=22)
plt.ylabel('Dollars', fontsize=22)
plt.title('Interest and Principal Paid per Month', fontsize=24)

# Number size Adjustment
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)


# Legends
plt.legend(loc=(1.02,0), borderaxespad=0, fontsize=20)

# SAVE IMAGE
plt.tight_layout()
plt.savefig('Files/carlegend.png', dpi = 200)
```
