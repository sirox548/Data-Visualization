## Removing NA or null values

```python
# Drop all NA or missing values in all the columns in the datass
df.dropna(how = 'any')
```

## Filiing in Missing Vaues

```python
# Fill NaN with zeros { if nan is recorded between row 30 to 39}
df['interest_paid'][30:40].fillna(0)
```

```python
# Back Fill in Values
df['interest_paid'][30:40].fillna(method='bfill')
```

```python
# Forward Fill in Values
df['interest_paid'][30:40].fillna(method='ffill')
```

```python
# Using Linear Interpolation (fill in the average values between the missing values in the NaN row or 0 value row
df['interest_paid'][30:40].interpolate(method='linear')
```
