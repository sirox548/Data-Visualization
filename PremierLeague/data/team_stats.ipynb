```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# path to read file
file_path = './teamstats-1920.csv'

# Read file
teamstats = pd.read_csv(file_path, index_col='Squad')
```

# line plot showing rate of goals score every 90 minutes per number of assist.

```python
plt.figure(figsize=(14,6))
plt.title('Premier league teams Goals per 90 Minutes')
sns.lineplot(data=teamstats['Gls Per 90 Minutes'], label='Gls Per 90 Minutes')
sns.lineplot(data=teamstats['Ast Per 90 Minutes'], label='Ast Per 90 Minutes')
plt.xlabel('Number of Goal per Assist.')
```

# bar plot showing rate of goals score every 90 minutes per number of assist.

```python
# width and height of chart
plt.figure(figsize=(14,10))

# Title
plt.title('Barplot showing the Yellow Card of each premier league team')
# barplot of teams that accumulated the most yellow cards
sns.barplot(y=teamstats.index, x=teamstats['CrdY'])
```

```python
# width and height of chart
plt.figure(figsize=(14,10))

# Title
plt.title('Barplot showing the Red Card of each premier league team')
# barplot of teams that accumulated the most yellow cards
sns.barplot(y=teamstats.index, x=teamstats['CrdR'])

### Arsenal are the most undisplined team with the most red card offense and yellow card offense
```
