Relational Plots

Scatter Plot (sns.scatterplot()):
Visualizes the relationship between two continuous variables.
Example: Scatter plot of height vs. weight.
Python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {'height': [170, 180, 165, 175, 190], 
        'weight': [70, 80, 65, 75, 90]}
df = pd.DataFrame(data)

# Create scatter plot
sns.scatterplot(x='height', y='weight', data=df)
plt.show()
Line Plot (sns.lineplot()):
Displays the relationship between two variables where one variable is continuous and the other is often time-based.
Example: Line plot of stock prices over time.
Python

# Assuming 'Date' is a column in your DataFrame
sns.lineplot(x='Date', y='StockPrice', data=df)
plt.show()
Joint Plot (sns.jointplot()):
Combines a scatter plot with histograms of the two variables on the axes.
Example: Joint plot of height and weight, showing their distributions.
Python

sns.jointplot(x='height', y='weight', data=df)
plt.show()
2. Categorical Plots

Bar Plot (sns.barplot()):
Displays the average or sum of a numerical variable across different categories.
Example: Bar plot of average sales per product category.
Python

# Assuming 'Category' and 'Sales' are columns in your DataFrame
sns.barplot(x='Category', y='Sales', data=df)
plt.show()
Count Plot (sns.countplot()):
Shows the frequency of occurrences for each category.
Example: Count plot of the number of customers in different cities.
Python

sns.countplot(x='City', data=df)
plt.show()
Box Plot (sns.boxplot()):
Displays the distribution of a numerical variable across different categories, showing the median, quartiles, and outliers.
Example: Box plot of salaries for different job titles.
Python

sns.boxplot(x='JobTitle', y='Salary', data=df)
plt.show()
Violin Plot (sns.violinplot()):
Similar to box plots, but provides a richer visualization of the data distribution, including the density of data points.
Example: Violin plot of salaries for different job titles.
Python

sns.violinplot(x='JobTitle', y='Salary', data=df)
plt.show()
Strip Plot (sns.stripplot()):
Shows the distribution of observations for each category.
Example: Strip plot of ages for different genders.
Python

sns.stripplot(x='Gender', y='Age', data=df)
plt.show()
Swarm Plot (sns.swarmplot()):
Similar to strip plot, but avoids overlapping points by adjusting their positions along the y-axis.
Example: Swarm plot of ages for different genders.
Python

sns.swarmplot(x='Gender', y='Age', data=df)
plt.show()
3. Distribution Plots

Histogram (sns.histplot()):
Displays the distribution of a single numerical variable.
Example: Histogram of ages in a population.
Python

sns.histplot(df['Age'], bins=20)
plt.show()
KDE Plot (sns.kdeplot()):
Plots the probability density function of a continuous variable.
Example: KDE plot of ages in a population.
Python

sns.kdeplot(df['Age'])
plt.show()
4. Matrix Plots

Heatmap (sns.heatmap()):
Visualizes a matrix of data, often used to represent correlations or other two-dimensional data.
Example: Heatmap of a correlation matrix.
Python

sns.heatmap(df.corr())
plt.show()
Pair Plot (sns.pairplot()):
Creates a grid of plots showing the relationships between all pairs of variables in a dataset.
Example: Pair plot of multiple variables in a dataset.
Python

sns.pairplot(df)
plt.show()
